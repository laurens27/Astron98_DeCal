```python
# Python DeCal
# Review Worksheet: NumPy Review

# Question 1: Fundamentals

# 1.1 Creating arrays
# Write some code to generate a 1D array of 1000 evenly spaced numbers from 0 to 2π. Then create a second
# array where each value in the array is double that of the previous array (call these x1 and x2 respectively).

import numpy as np

x1 = np.linspace(0, 2*np.pi, 1000)
x2 = 2 * x1


# 1.2 Statistics with NumPy
# Take the mean, median, and standard deviation of x1 and x2 using NumPy.

# x1
mean_x1 = np.mean(x1)
median_x1 = np.median(x1)
std_dev_x1 = np.std(x1)

# x2
mean_x2 = np.mean(x2)
median_x2 = np.median(x2)
std_dev_x2 = np.std(x2)


# 1.3 Slicing and Striding
# Create a 2D array called filter of shape (200, 100) populated with only 0’s. Then set the second quadrant of
# this equal to 1’s using index slicing, and the fourth quadrant equal to 2. Hint (The first 500 indices in each
# axis is what we want to set = 1)

filter = np.zeros((200, 100))

filter[:100, :100] = 1 

filter[100:, :100] = 2 


# 1.3.1 Axis-Wise Operations
# Using the array from Section 1.3, perform the same statistical operations from section 1.2 across the rows
# and then across the columns of each. You will know if you have done the right axis if the length of the array
# matches the shape axis we specified above.

mean_rows = np.mean(filter, axis=1)
mean_cols = np.mean(filter, axis=0)

median_rows = np.median(filter, axis=1)
median_cols = np.median(filter, axis=0)

std_dev_rows = np.std(filter, axis=1)
std_dev_cols = np.std(filter, axis=0)
```

```python
# Question 2: Boolean Filtering
# Create an array of 100 numbers using np.linspace or np.arange from 0 to 100. Use boolean filtering to
# create new arrays from the one we just created that satisfy the following criteria:
# • Only even numbers.
# • Only odd numbers that are divisible by 7.
# • Numbers divisible by 10 and divisible by 5.
# • Numbers that are greater than 42.
# • Only numbers that are multiples of 2,3, and 5.

numbers = np.arange(101)

even_numbers = numbers[numbers % 2 == 0]

odd_divisible_by_7 = numbers[(numbers % 2 != 0) & (numbers % 7 == 0)]

divisible_by_10_and_5 = numbers[(numbers % 10 == 0) & (numbers % 5 == 0)]

greater_than_42 = numbers[numbers > 42]
    
multiples_of_2_3_5 = numbers[(numbers % 2 == 0) & (numbers % 3 == 0) & (numbers % 5 == 0)]
```
