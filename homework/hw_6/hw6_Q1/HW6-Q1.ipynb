```python
# Python DeCal
# HW 6: Plotting
# Question 1 - Random Plots

# This problem is an exercise in plotting random data. First you must generate three lists of random numbers,
# each 40 elements long with values ranging from 0 to 100. Then make two stacked vertical subplots. 
# The top plot should plot one list of random data in orange 
# and with a line width of 10 and another list in red and dashed. 
# The bottom plot should be a scatter plot of the final list with magenta diamonds as the marker. 
# In each case the x coordinate should be the position in the list. 

import matplotlib.pyplot as plt
import numpy as np

# generate random number lists
list1 = np.random.randint(0, 101, size=40)
list2 = np.random.randint(0, 101, size=40)
list3 = np.random.randint(0, 101, size=40)

plt.figure(figsize=(8, 6))

# Top Plot
plt.subplot(2, 1, 1)
plt.plot(range(len(list1)), list1, color='orange', linewidth=10, label='List 1')
plt.plot(range(len(list2)), list2, color='red', linestyle='--', label='List 2')
plt.title('Top Plot')
plt.legend()

# Bottom Plot
plt.subplot(2, 1, 2)
plt.scatter(range(len(list3)), list3, color='magenta', marker='D', label='List 3')
plt.title('Bottom Plot')
plt.legend()

plt.tight_layout()
plt.show()
```
