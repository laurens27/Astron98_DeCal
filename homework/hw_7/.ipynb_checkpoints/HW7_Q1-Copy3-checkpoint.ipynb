```python
# Python DeCal
# HW 7: Data with Pandas
# Question 1 - Tweet Analysis

import pandas as pd

# using json rather than csv because my computer couldn't process csv
file_path = 'AOC_recent_tweets.txt'
data = pd.read_json(file_path).set_index('id')

def time_in_hours(created_at):
    created_at = pd.to_datetime(created_at)
    decimal_hours = created_at.dt.hour + created_at.dt.minute / 60 + created_at.dt.second / (60**2)
    return decimal_hours

data['hours'] = time_in_hours(data['created_at'])

selected_columns = ['created_at', 'hours', 'full_text']

selected_data = data[selected_columns]

output_file_path = 'AOC_organized_tweets.csv'
selected_data.to_csv(output_file_path)
```
